```python
import matplotlib
%matplotlib inline
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import sys
from layers import nms,iou

from glob import glob
import random
```

```python
img_path = sorted(glob('/home/432/ysw/EGFR/egfr_prep_result/*_clean.npy'))
pbb_path = sorted(glob('/home/432/ysw/DSB2017/bbox_result_egrf/*_pbb.npy'))
```

```python
for img_p, pbb_p in random.sample(list(zip(img_path, pbb_path)), 20):
    img = np.load(img_p)
    pbb = np.load(pbb_p)
    
    pbb = pbb[pbb[:,0]>-1]

    pbb = nms(pbb,0.05)
    box = pbb[0].astype('int')[1:]
    
    print(img_p.split('/')[-1][:-10])
    ax = plt.subplot(1,1,1)
    plt.imshow(img[0,box[0]],'gray')
    plt.axis('off')
    rect = patches.Rectangle((box[2]-box[3],box[1]-box[3]),box[3]*2,box[3]*2,linewidth=2,edgecolor='red',facecolor='none')
    ax.add_patch(rect)
    plt.show()
```
